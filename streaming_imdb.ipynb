```python
import re

def bersihkan_siku_pangkat(judul):
    # Cari pola seperti [REC²], [HD], dst
    matches = re.findall(r"\[(.*?)\]", judul)
    for match in matches:
        # Cek apakah ada karakter pangkat di dalamnya
        pangkat = ''.join(c for c in match if c in '²³¹⁰⁴⁵⁶⁷⁸⁹⁺⁻')
        teks = match.replace(pangkat, "")
        if pangkat:
            pengganti = f"{pangkat}{teks}"
        else:
            pengganti = teks
        # Ganti dalam judul: hilangkan kurung siku dan atur ulang
        judul = judul.replace(f"[{match}]", pengganti)
    return judul.strip()
```

```python
def keep_only_names(cast_entry):
    try:
        if isinstance(cast_entry, str):
            import json
            cast_entry = json.loads(cast_entry)

        if isinstance(cast_entry, list):
            # Jika sudah list of string (hasil faker), langsung kembalikan
            if all(isinstance(c, str) for c in cast_entry):
                return cast_entry
            # Jika list of dict, ambil hanya 'name'
            elif all(isinstance(c, dict) for c in cast_entry):
                return [c.get("name", "") for c in cast_entry]
    except:
        return []
    
    return []
```

```python
from neo4j import GraphDatabase
from tqdm import tqdm

driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "hello123"))

def insert_to_neo4j(tx, title, genres, cast):
    tx.run("MERGE (m:Movie {title: $title})", title=title)

    for g in genres:
        tx.run("MERGE (g:Genre {name: $name})", name=g) 
        tx.run("""
            MATCH (m:Movie {title: $title}), (g:Genre {name: $name})
            MERGE (m)-[:BELONGS_TO]->(g)
        """, title=title, name=g)

    for actor in cast[:10]:  # Batasin 5 aktor
        tx.run("MERGE (a:Actor {name: $name})", name=actor)
        tx.run("""
            MATCH (m:Movie {title: $title}), (a:Actor {name: $name})
            MERGE (a)-[:ACTED_IN]->(m)
        """, title=title, name=actor)

# Proses Upload
with driver.session() as session:
    for idx, row in tqdm(df_neo4j.iterrows(), total=len(df_neo4j)):
        try:
            title = row["title"]
            genres = row["genres"] if isinstance(row["genres"], list) else []
            cast = row["cast"] if isinstance(row["cast"], list) else []

            session.write_transaction(insert_to_neo4j, title, genres, cast)
        except Exception as e:
            print(f"❌ Error on row {idx}: {e}")
            continue
```

## Upload ArangoDB

```python
from arango import ArangoClient
from arango.exceptions import ArangoServerError
from tqdm import tqdm

def upload_to_arango(df_arango):
    client = ArangoClient()
    db = client.db("film_db", username="root", password="tes")

    if not db.has_collection("moviess"):
        db.create_collection("moviess")
    col = db.collection("moviess")

    failed_docs = []

    for _, row in tqdm(df_arango.iterrows(), total=len(df_arango)):
        try:
            doc = row.dropna().to_dict()  # Hindari nilai NaN
            doc["_key"] = str(row["id"])  # Gunakan id sebagai key unik
            col.insert(doc, overwrite=True)
        except ArangoServerError as e:
            failed_docs.append(doc)  # Simpan yang gagal untuk diulang
            print(f"❌ Gagal upload ID {row['id']}: {e}")

    print(f"\n📌 Total gagal upload: {len(failed_docs)} dokumen")

    return failed_docs  # Bisa dipakai untuk upload ulang jika mau
```

```python
import time
from neo4j import GraphDatabase

driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "hello123"))

def measure_query(query, desc=""):
    with driver.session() as session:
        start = time.perf_counter()
        result = session.run(query).data()
        end = time.perf_counter()
        print(f"⏱️ {desc} - Waktu: {end - start:.4f} detik")
        return result, end - start

query = """
MATCH (m:Movie)
WHERE m.title STARTS WITH 'A'
RETURN m.title LIMIT 50
"""

# Belum pakai index
result_before, time_before = measure_query(query, "Sebelum Indexing")

# Buat index dulu
with driver.session() as session:
    session.run("CREATE INDEX FOR (m:Movie) ON (m.title)")

# Pakai index
result_after, time_after = measure_query(query, "Sesudah Indexing")
```

```python
import time
from neo4j.exceptions import Neo4jError

def compare_neo4j_query_before_after_indexing(session, query, create_index_query=None, label="Query Neo4j"):
    print(f"🔍 Membandingkan performa: {label}\n")

    try:
        # 🔹 Sebelum indexing
        start_before = time.time()
        result_before = list(session.run(query))
        end_before = time.time()
        time_before = end_before - start_before
        print(f"⏳ Sebelum indexing: {time_before:.4f} detik | Hasil: {len(result_before)}")

        # 🔹 Tambahkan indexing jika disediakan
        if create_index_query:
            print("⚙️  Membuat index...")
            session.run(create_index_query)
            print("✅ Index berhasil dibuat.\n")

        # 🔹 Sesudah indexing
        start_after = time.time()
        result_after = list(session.run(query))
        end_after = time.time()
        time_after = end_after - start_after
        print(f"🚀 Sesudah indexing: {time_after:.4f} detik | Hasil: {len(result_after)}")

        # 🔚 Perbandingan
        improvement = time_before - time_after
        print(f"\n📊 Perbaikan performa: {improvement:.4f} detik lebih cepat\n")

    except Neo4jError as e:
        print(f"❌ Error saat menjalankan query: {e}")
```

```python
import time

def compare_arango_query_before_after_indexing(db, query, create_index_fn=None, label="Query Arango"):
    print(f"🔍 Membandingkan performa: {label}\n")

    # 🔹 Sebelum indexing
    start_before = time.time()
    cursor_before = db.aql.execute(query)
    result_before = list(cursor_before)
    end_before = time.time()
    time_before = end_before - start_before
    print(f"⏳ Sebelum indexing: {time_before:.4f} detik | Hasil: {len(result_before)}")

    # 🔹 Buat index jika disediakan
    if create_index_fn:
        print("⚙️  Membuat index...")
        create_index_fn()
        print("✅ Index dibuat.\n")

    # 🔹 Sesudah indexing
    start_after = time.time()
    cursor_after = db.aql.execute(query)
    result_after = list(cursor_after)
    end_after = time.time()
    time_after = end_after - start_after
    print(f"🚀 Sesudah indexing: {time_after:.4f} detik | Hasil: {len(result_after)}")

    # 🔚 Perbandingan
    improvement = time_before - time_after
    print(f"\n📊 Perbaikan performa: {improvement:.4f} detik lebih cepat\n")
```

```python
def create_index():
    collection = db.collection("movies")
    # Cek apakah index sudah ada
    existing_indexes = collection.indexes()
    already_indexed = any(
        index["type"] == "persistent" and index["fields"] == ["vote_average"]
        for index in existing_indexes
    )
    if not already_indexed:
        collection.add_persistent_index(fields=["vote_average"])
```
